```python
#Q4, Grade Calculation - Implement a method called calculate_grade() in the Student class that calculates the grade based on the calculated percentage. Grading scale of the institute is given below:
class Student:
    def __init__(self, enrollment_ID, student_name, course_name, age, marks, grade=None):
        self.enrollment_ID = enrollment_ID
        self.student_name = student_name
        self.course_name = course_name
        self.age = age
        self.marks = marks
        self.grade = grade

        self.age_validation()
        self.percentage = self.validate_and_convert_marks()
        self.grade = self.calculate_grade()

    def age_validation(self):
        if not isinstance(self.age, int) or self.age < 3 or self.age > 60:
            raise ValueError(f"Invalid age: {self.age}. Age must be between 3 and 60.")

    def validate_and_convert_marks(self):
        if 0 <= self.marks <= 44:
            return (self.marks / 44) * 100
        else:
            return 'Invalid Marks'

    def calculate_grade(self):
        if isinstance(self.percentage, str):
            return 'Invalid Marks' if self.percentage == 'Invalid Marks' else 'Impossible'
        
        if self.percentage > 100:
            return 'Impossible'
        elif self.percentage > 80:
            return 'Grade A'
        elif self.percentage > 60:
            return 'Grade B'
        elif self.percentage > 40:
            return 'Grade C'
        elif self.percentage > 20:
            return 'Grade D'
        elif self.percentage >= 0:
            return 'Grade F'

    def __str__(self):
        return (f"Enrollment ID: {self.enrollment_ID}\n"
                f"Name: {self.student_name}\n"
                f"Course: {self.course_name}\n"
                f"Age: {self.age}\n"
                f"Marks: {self.marks}\n"
                f"Percentage: {self.percentage}%\n"
                f"Grade: {self.grade}")

# Example usage:
try:
    student1 = Student(
        enrollment_ID="S12345",
        student_name="John Doe",
        course_name="Computer Science",
        age=20,
        marks=30
    )
    print(student1)

    student2 = Student(
        enrollment_ID="S67890",
        student_name="Jane Smith",
        course_name="Mathematics",
        age=22,
        marks=50  # Invalid marks
    )
    print(student2)

except ValueError as e:
    print(e)
```

```python
#Q6a. The constructor should initialize the enrollment ID attribute, prompting the user to input an integer enrollment ID.
class Guardian:
    def __init__(self):
        while True:
            try:
                self.enrollment_ID = int(input("Enter the enrollment ID (integer): "))
                if self.enrollment_ID <= 0:
                    raise ValueError("Enrollment ID must be a positive integer.")
                break
            except ValueError as e:
                print(e)
                print("Please enter a valid integer.")

guardian1 = Guardian()
print(f"Guardian Enrollment ID: {guardian1.enrollment_ID}")
```

```python
#Q6b. Implement a class method named 'add_guardian()' to facilitate adding guardian details. This method should prompt the user to enter the guardian's name, update the total number of guardians managed by the class, and provide an option to add additional guardians. It should then print the details of all guardians entered.
class Guardian:
    total_guardians = 0  
    @classmethod
    def add_guardian(cls):
        guardians = []
        while True:
            guardian_name = input("Enter the guardian's name: ")
            guardians.append(guardian_name)
            cls.total_guardians += 1

            more = input("Do you want to add another guardian? (yes/no): ").strip().lower()
            if more != 'yes':
                break

        print("\nAll Guardian Details:")
        for idx, name in enumerate(guardians, 1):
            print(f"Guardian {idx}: {name}")

        print(f"Total number of guardians managed: {cls.total_guardians}")

# Example usage:
guardian1 = Guardian()
Guardian.add_guardian()
```
