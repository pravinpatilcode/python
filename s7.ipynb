```python
#                               sprint7
#Q1, Create the Product interface with an abstract method calculate_total_price.
from abc import ABC, abstractmethod

class Product(ABC):
    @abstractmethod
    def calculate_total_price(self):
        pass

# Example of a concrete class implementing the Product interface
class ConcreteProduct(Product):
    def __init__(self, price, quantity):
        self.price = price
        self.quantity = quantity

    def calculate_total_price(self):
        return self.price * self.quantity

# Example usage:
product = ConcreteProduct(price=100, quantity=5)
print(f"Total Price: ${product.calculate_total_price()}")
```

```python
#Q2, Create concrete subclass Digital Product with the attributes such as product_name, price, weight, shipping_charge, delivery_charge, and its own implementation of calculate_total_price().
from abc import ABC, abstractmethod

class Product(ABC):

    @abstractmethod
    def calculate_total_price(self):
        pass

class DigitalProduct(Product):
    def __init__(self, product_name, price, weight, shipping_charge, delivery_charge):
        self.product_name = product_name
        self.price = price
        self.weight = weight
        self.shipping_charge = shipping_charge
        self.delivery_charge = delivery_charge

    def calculate_total_price(self):
        return self.price + self.shipping_charge + self.delivery_charge

    def __str__(self):
        return (f"Product Name: {self.product_name}\n"
                f"Price: ${self.price}\n"
                f"Weight: {self.weight}kg\n"
                f"Shipping Charge: ${self.shipping_charge}\n"
                f"Delivery Charge: ${self.delivery_charge}\n"
                f"Total Price: ${self.calculate_total_price()}")

# Example usage:
digital_product = DigitalProduct(
    product_name="E-Book",
    price=29.99,
    weight=0.5,
    shipping_charge=0,  # Digital products might not have shipping charges
    delivery_charge=0
)

print(digital_product)
```

```python
#Q4, Create the Order class with methods to add items, get customer name, get items, and get the total price. Encapsulate customer name, items, and total price with appropriate data protection.Ensure that the add_item method updates the order's items list and total price accurately.
class Order:
    def __init__(self, customer_name):
        self._customer_name = customer_name 
        self._items = [] 
        self._total_price = 0.0  

    def add_item(self, item_name, item_price):
        self._items.append((item_name, item_price))
        self._total_price += item_price

    def get_customer_name(self):
        return self._customer_name

    def get_items(self):
        return self._items

    def get_total_price(self):
        return self._total_price

# Example usage:
order = Order("John Doe")
order.add_item("Laptop", 999.99)
order.add_item("Mouse", 25.50)
order.add_item("Keyboard", 45.00)

print(f"Customer Name: {order.get_customer_name()}")
print(f"Items: {order.get_items()}")
print(f"Total Price: ${order.get_total_price():.2f}")
```

```python
#Q5, Create a main function that simulates the process of placing orders in the online store.Instantiate products of different types and add them to orders.Display the customer's name, items in the order, and the total price for each order placed.Test the main function with various scenarios to ensure the system handles orders correctly.
from abc import ABC, abstractmethod

# Define the Product interface
class Product(ABC):
    @abstractmethod
    def calculate_total_price(self):
        pass

# Define the DigitalProduct class
class DigitalProduct(Product):
    def __init__(self, product_name, price, weight, shipping_charge, delivery_charge):
        self.product_name = product_name
        self.price = price
        self.weight = weight
        self.shipping_charge = shipping_charge
        self.delivery_charge = delivery_charge

    def calculate_total_price(self):
        return self.price + self.shipping_charge + self.delivery_charge

    def __str__(self):
        return (f"Product Name: {self.product_name}\n"
                f"Price: ${self.price}\n"
                f"Weight: {self.weight}kg\n"
                f"Shipping Charge: ${self.shipping_charge}\n"
                f"Delivery Charge: ${self.delivery_charge}\n"
                f"Total Price: ${self.calculate_total_price()}")

# Define the PhysicalProduct class
class PhysicalProduct(Product):
    def __init__(self, product_name, price, weight, shipping_charge, delivery_charge):
        self.product_name = product_name
        self.price = price
        self.weight = weight
        self.shipping_charge = shipping_charge
        self.delivery_charge = delivery_charge

    def calculate_total_price(self):
        return self.price + self.shipping_charge + self.delivery_charge

    def __str__(self):
        return (f"Product Name: {self.product_name}\n"
                f"Price: ${self.price}\n"
                f"Weight: {self.weight}kg\n"
                f"Shipping Charge: ${self.shipping_charge}\n"
                f"Delivery Charge: ${self.delivery_charge}\n"
                f"Total Price: ${self.calculate_total_price()}")

# Define the Order class
class Order:
    def __init__(self, customer_name):
        self._customer_name = customer_name
        self._items = []
        self._total_price = 0.0

    def add_item(self, item):
        self._items.append(item)
        self._total_price += item.calculate_total_price()

    def get_customer_name(self):
        return self._customer_name

    def get_items(self):
        return self._items

    def get_total_price(self):
        return self._total_price

# Main function to simulate the process of placing orders
def main():
    # Create products
    digital_product = DigitalProduct(
        product_name="E-Book",
        price=12.99,
        weight=0,
        shipping_charge=0,
        delivery_charge=0
    )

    physical_product = PhysicalProduct(
        product_name="Smartphone",
        price=299.99,
        weight=0.5,
        shipping_charge=10.00,
        delivery_charge=15.00
    )

    order1 = Order("Alice Johnson")
    order1.add_item(digital_product)
    order1.add_item(physical_product)

    order2 = Order("Bob Smith")
    order2.add_item(physical_product)  # Adding only physical product to this order

    def display_order_details(order):
        print(f"Customer Name: {order.get_customer_name()}")
        print("Items in the Order:")
        for item in order.get_items():
            print(f" - {item.product_name}: ${item.calculate_total_price()}")
        print(f"Total Price: ${order.get_total_price():.2f}")
        print("-" * 40)

    print("Order 1 Details:")
    display_order_details(order1)

    print("Order 2 Details:")
    display_order_details(order2)

# Run the main function to test the system
if __name__ == "__main__":
    main()
```
